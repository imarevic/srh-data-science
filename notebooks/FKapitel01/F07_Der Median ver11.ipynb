## Der Median
----------------------------------------

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as st
```

Ein weiteres sehr wichtiges Maß für die zentrale Tendenz ist der **Median**. Der Median von Meßwerten ist der Wert des mittleren Terms in einem Datensatz, der in aufsteigender Reihenfolge geordnet wurde. Der Median unterteilt also einen geordneten Datensatz **in zwei gleiche Teile**. <br> <br> Die Berechnung des Medians besteht aus den folgenden zwei Schritten: <br> <br> 1.Ordnen Sie den Datensatz in aufsteigender Reihenfolge.
    
2.Finden Sie den mittleren Term. Der Wert dieses Terms ist der Median.

Wenn die Anzahl der Beobachtungen in einem Datensatz ungerade ist, wird der Median durch den Wert des mittleren Terms in den geordneten Daten bestimmt. Ist die Anzahl der Beobachtungen jedoch gerade, so wird der Median durch den Durchschnitt der Werte der beiden mittleren Terme bestimmt *(Mann 2012)*. <br> <br> Lassen Sie uns den Median für die Variable `age` des `students` Datensatzes auswerten.

```python
# Einlesen der Datei students.csv
students=pd.read_csv('students.csv', index_col=0)

# Lese Spalte der Werte für Alter in Variable age ein
age = students['age']

# Erzeuge Achsenbeschriftungen
plt.xlabel('Index',  
           color='k',
           labelpad = 6)
plt.ylabel('Alter der Studenten',  
           color='k', 
           labelpad = 6)

# Plotten der Werte für Alter
plt.plot(age,'o',markeredgecolor='k',markerfacecolor='None')
plt.show()
```

Wenn wir die Variable `age` plotten, erkennen wir sofort, dass es einige Studenten gibt, die viel älter sind als der Rest der Studenten. Lassen Sie uns den Median berechnen...

```python
median_alter = st.median(age)
print('Median des Alters der Studenten :' , median_alter)
```

... und vergleichen wir es mit dem arithmetischen Mittel.

```python
mean_alter = np.mean(age)
print('Mittelwert des Alters der Studenten :' , mean_alter)
```

Zur Veranschaulichung fügen wir nun den Median und das arithmetische Mittel in das Diagramm ein

```python

# Plotte Werte für Alter
fig, ax = plt.subplots()
plt.plot(age,'o',markeredgecolor='k',markerfacecolor='None')

# Zeichne Linien bei Wert für Median(grün) und Mittelwert(rot)
plt.axhline(y= median_alter, xmin= 0.0,xmax= 1.0,color= 'g')
plt.axhline(y= mean_alter, xmin= 0.0,xmax= 1.0,color= 'r')

# Erzeuge Achsenbeschriftungen
plt.xlabel('Index',  
           color='k',
           labelpad = 6)
plt.ylabel('Alter der Studenten',  
           color='k', 
           labelpad = 6)

# Dargestellter Bereich der y-Achse
plt.ylim(0,85)

# Erzeuge Legende
ax.plot(median_alter, 'g--', label='Median')
ax.plot(mean_alter, 'r--', label='Mittelwert')
legend = ax.legend(loc='upper right', shadow=True, fontsize='small')


plt.show()
```

Der Median wird durch Ausreißer nicht beeinflusst. Daher wird der Median als Maß für die zentrale Tendenz bei Datensätzen, die Ausreißer enthalten, dem Mittelwert vorgezogen (Mann 2012).
