## Der Median
----------------------------------------

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as st
```

Ein weiteres sehr wichtiges Maß für die zentrale Tendenz ist der **Median**. Der Median von Meßwerten ist der Wert des mittleren Terms in einem Datensatz, der in aufsteigender Reihenfolge geordnet wurde. Der Median unterteilt also einen geordneten Datensatz **in zwei gleiche Teile**. <br> <br> Die Berechnung des Medians besteht aus den folgenden zwei Schritten: <br> <br> 1.Ordnen Sie den Datensatz in aufsteigender Reihenfolge.
    
2.Finden Sie den mittleren Term. Der Wert dieses Terms ist der Median.

Wenn die Anzahl der Beobachtungen in einem Datensatz ungerade ist, wird der Median durch den Wert des mittleren Terms in den geordneten Daten bestimmt. Ist die Anzahl der Beobachtungen jedoch gerade, so wird der Median durch den Durchschnitt der Werte der beiden mittleren Terme bestimmt ({cite:p}`fahrmeirstatistik` s.52). <br> <br> Lassen Sie uns den Median für die Variable `age` des `students`datensatzes auswerten.

```python
df = pd.read_csv("students.csv", index_col=0)
df.plot.scatter(x="stud.id", y="age")
```

```python
df=pd.read_csv('students.csv', index_col=0)

age = df['age']
plt.xlabel('Index')
plt.ylabel('Alter')
plt.plot(age,'ro')
plt.show()
```

Wenn wir die Variable `age` plotten, erkennen wir sofort, dass es einige Studenten gibt, die viel älter sind als der Rest der Studenten. Lassen Sie uns den Median berechnen...

```python
median_alter = st.median(df.age)
mean_alter = np.mean(df.age)
print(f"Median des Alters der Studenten: {median_alter}")
print(f"Mittelwert des Alters der Studenten: {mean_alter}")
```

Zur Veranschaulichung fügen wir nun den Median und das arithmetische Mittel in das Diagramm ein

```python
plt.plot(age,'ro',median_alter, 'bs', mean_alter , 'g^')
plt.ylim(15,85)
plt.axhline(y= median_alter, xmin= 0.0,xmax= 1.0,color= 'b', label = 'Median')
plt.axhline(y= mean_alter, xmin= 0.0,xmax= 1.0,color= 'g',  label = 'arithmetisches Mittel')
plt.legend()
plt.show()
```

Der Median wird durch Ausreißer nicht beeinflusst. Daher wird der Median als Maß für die zentrale Tendenz bei Datensätzen, die Ausreißer enthalten, dem Mittelwert vorgezogen ({cite:p}`fahrmeirstatistik` s.54).
